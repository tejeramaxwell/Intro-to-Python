```python
#Question 2: What is the longest name in the list from "names.txt"?
names = open('names.txt','r') #opening file and converting to variable
max_count = 0 #setting original count for characters
max_name = '' #setting the original name of maximum
for line in names : #iterating through lines of file
    name = line.rstrip()
    character_count = len(name) #counting characters in the file
    if character_count > max_count: #if the character count of the lines is greater than the prior maximum count, it becomes the new maximum
        max_count = character_count #the new maximum becomes the character count
        max_name = str(name) #the new maximum name becomes the name

print('Longest Name = ',max_name)
print('Longest Name Character Count =',max_count,'characters')

#This works, but it bothers me that it doesn't use the 'max' function. That sounds more intuitive.
```

```python
#Question 3: What is the shortest name in the list from "names.txt"?

#Let's try this one with the min function and a dictionary to track our data.

names = open('names.txt','r') #converts text file to string
names_dict = dict() #starts a dictonary
for line in names : #iterates through lines
    name = line.rstrip() #drops line separator character
    name_len = len(name) #counts the length of the line / name
    names_dict[name] = name_len #adds the name and the count to the dictionary

min_name_count = min(names_dict.values()) #identifies the minimum character count in dictionary
min_name_str = [key for key in names_dict if names_dict[key] == min_name_count] 
#identifies the names in the dictionary with a character count equal to the minimum character count
print("Names with minimum value are:",min_name_str,"With a character count of:",min_name_count)
#prints the keys with a character count equal to the minimum character count and the minimum character count
```

```python
#Question 4: What is the average length (rounded to two decimal places) of the names in the list from "names.txt"?
 
```

```python
#Now that we have a dictionary, this should be easier as we can just substitute min() for some averaging code
names = open('names.txt','r') #converts text file to string
names_dict = dict() #starts a dictonary
count_names_dict = 0 #starts the count of names in dictionary
for line in names : #iterates through lines
    name = line.rstrip() #drops line separator character
    name_len = len(name) #counts the length of the line / name
    names_dict[name] = name_len #adds the name and the character count to the dictionary
    count_names_dict = count_names_dict + 1 #increases the count of names in the dictionary

mean_name_count = round(sum(names_dict.values())/count_names_dict,2) #identifies the mean character count in dictionary
print("The average length of names is:",mean_name_count,"characters")
```

```python
#Question 8: Find and display the county with the highest population.
#This has an obvious result - Denver county.
#Let's try this without looking at the underlying data before running our script.
```

```python
census = open('Colorado_county_race.json')
census = json.load(census)
max_name = "" #set original county name to nothing
pop_max = 0 #set the original max population to 0
for row in census : #iterate between rows, which we know are each county
    if (row['Total population']) > pop_max: #checks if the county we're iterating through has a higher population than the maximum
        pop_max = (row['Total population']) #replaces the maximum if so
        max_name = (row['County']) #holds the name of the maximum
    
print("The most populous county is",max_name,"with population:",pop_max) #show the name of the most populous county and its population
```

```python
#Question 9: #3. Find and display the county with the lowest population.
#This has a less obvious result, but should be a pretty simple coding process.
#We could do the same thing as before swapping the greater than with the less than sign, but let's get more sophisiticated
#like we did earlier in this problem set. Let's use the min function and a new dictionary
```

```python
census = open('Colorado_county_race.json')
census = json.load(census)    

min_population = min(([row['Total population'] for row in census]))

for row in census:
    if row['Total population'] == min_population:
        print('Lowest population county:',row['County'],", population:",min_population)
```

```python
#Question 10: Calculate and display the average population of the counties in Colorado (rounded to the nearest integer).

census = open('Colorado_county_race.json')
census = json.load(census)    

total_pop = sum(([row['Total population'] for row in census])) #find total population
county_count = 0 #start county of counties
for row in census:
    county_count = county_count + 1 #count the counties
avg_county_pop = total_pop / county_count #find the average
print('Average county population:',int(round(avg_county_pop,0))) #display as a rounded integer
```
