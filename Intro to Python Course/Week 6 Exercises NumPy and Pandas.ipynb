```python
#Practice Problem 1:
#You are working with a 2D NumPy array of integers, where each element represents the elevation of a corresponding area
#on a map. You need to transpose this array to view the landscape from a different perspective.
#Task:
#Create a 2D NumPy array: This array should represent a grid of elevations, with each integer value ranging from 1 to 1000. 
#The shape of the array should be (5, 5).
#Transpose the Array: Write a function that takes in the 2D NumPy array and returns a new array that is a transposition 
#of the original array. The transposition should be done using NumPy functionalities.
#Display the Original and Transposed Array: Finally, print both the original and the transposed array to the console.
```

```python
#Import NumPy module, as np so future commands can be more concise:
import numpy as np

#Use the np.array() command to build a 5x5 array with arbitrary values ranging from 1 to 1000
MAP=np.array([[100,250,500,750,900],[50,150,350,600,800],[75,175,275,375,475],[100,200,300,400,500],[1000,900,800,700,600]])

#Use a new variable and the .transpose() function to "flip" the orientation of the array
MAP_transpose = MAP.transpose()

#Show results
print(MAP,'\n')
print(MAP_transpose)
```

```python
#Practice Problem 2:
#You have been given temperature data recorded in a city over a week. The data is stored in a 2D NumPy array, 
#where each row represents a day, and each column represents an hourly temperature reading in Celsius.
#Task:
#Initialize the Temperature Data Array: Create a 2D NumPy array with 7 rows (for 7 days) and 24 columns (for 24 hours). 
#Fill this array with random integers representing the temperature readings between 15 and 35 degrees Celsius.
#Extract Morning Temperatures: Write a function that takes in the temperature data array and returns a new array 
#containing only the temperature readings from 6 AM to 9 AM for every day.
#Calculate Average Morning Temperature: Write a function to calculate the average morning temperature over the week
#using the sliced array from the previous step.
#Display Results: Print the extracted morning temperatures and the average morning temperature to the console.
```

```python
#First lets generate our set of temperatures

#Specify temperature range and dimensions of the array
low_temp = 15
high_temp = 35
temp_array_shape = (7, 24)

#Use the NP randomint function to generate random values based on our specifications
temp_array = np.random.randint(low_temp,high_temp,size=temp_array_shape)
print('Random Temperatures for each hour throughout the week:\n',temp_array,'\n')

#Specify morning hours

morning_temp_array=temp_array[:,[6,7,8,9]]
print('Morning temperatures:\n',morning_temp_array)

#Calculate average temp
average_morning_temp_array = np.mean(morning_temp_array)
print('The average morning temperature is:',average_morning_temp_array)
```

```python
#Practice Problem 3:
#Imagine you are analyzing a dataset representing the ages and heights of a group of people. 
#You are interested in filtering out the data to focus on specific age groups and height ranges.
#Task:
#Create Age and Height Arrays: Generate two 1D NumPy arrays â€“ one representing the ages of 100 people 
#(integer values between 18 and 60) and the other representing the corresponding heights in centimeters 
#(float values between 150.0 and 200.0).
#Filter Data Using Boolean Indexing: Use boolean indexing to filter the arrays and obtain the heights of people who are:
#Between 25 and 35 years old, inclusive.
#Taller than 170 cm.
#Calculate Average Height: After filtering, calculate the average height of the people who meet the above criteria.
#Output: Display the filtered heights and their average.
```

```python
#Define row count for each vector
size_a = (100)
size_h = (100)

#Set range for random values to be generated within vectors
min_a = 18
max_a = 60
min_h = 150.0
max_h = 200.0

#Run separate random functions for integers and floats to produce two different arrays
column_a = np.random.randint(min_a,max_a,size_a)
column_h = np.random.uniform(min_h,max_h,size_h)

#Merge the arrays into a 2x100 matrix
array_ah = np.column_stack((column_a,column_h))



#Set filter criteria
filter_a_min = 25
filter_a_max = 35

#Create new array based on filter criteria, note the >= and <= to be inclusive
filtered_array = array_ah[(array_ah[:, 0] >= filter_a_min) & (array_ah[:, 0] <= filter_a_max)]

#Create a vector based on filtered matrix array
filtered_h = filtered_array[:, 1]

#Calculate mean of the filtered vector
mean_filtered_h = np.mean(filtered_h)

#Show results
print('The whole data set is:\n',array_ah)
print('The data set filtered to only ages between 25 and 35 inclusive is:\n',filtered_array)
print('The mean height of people in the dataset with ages between 25 and 35 is:\n',mean_filtered_h)
```

```python
#Practice Problem 7:
#With your home sales DataFrame, you are tasked with analyzing the dataset to understand the central tendencies
#and spread of specific variables. This will help in gaining insights into the general landscape of home sales in the dataset.
#Task:
#Price Analysis:
#Calculate and display the mean, median, mode, and standard deviation of the price column.
#Identify and display the highest and lowest price in the dataset.
#Square Feet Analysis:
#Calculate and display the mean, median, mode, and standard deviation of the square feet column.
#Identify and display the largest and smallest home in terms of square footage.
#Year Built Analysis:
#Calculate and display the mean, median, and mode of the year built column.
#Identify and display the oldest and newest home in the dataset.
#Bedrooms and Bathrooms Analysis:
#Calculate and display the average number of bedrooms and bathrooms in the dataset.
#Identify and display the home with the most bedrooms and the home with the most bathrooms.
#Price Per Square Foot:
#Create a new column price per square foot in the DataFrame.
#Calculate and display the mean, median, mode, and standard deviation of this new column.
```

```python
#Price, sqft, and year built analysis
#Run a loop on a list to calculate summary statistics for three variables of interest
#Use the mean, median, mode, std, min, and max functions to calculate and print summary statistics.

variables = ['price','sqft_lot','yr_built']

for var in variables :
    mean = df[var].mean()
    median = df[var].median()
    mode = df[var].mode()
    std_dev = df[var].std()
    minimum=df[var].min()
    maximum=df[var].max()
    print("Variable name:",var)
    print("Mean:", mean)
    print("Median:", median)
    print("Mode:", mode)
    print("Standard Deviation:", std_dev)
    print("Minimum:", minimum)
    print("Maximum:", maximum)
    print("\n")

```

```python
#Identify and display the largest and smallest homes.
#Identify and display the oldest and newst home in the dataset.
#Identify and display the home with the most bedrooms and the home with the most bathrooms.

print('Home(s) with the most square footage lot:')

#Display data for homes with the square footage equal to the maximum square footage:
print(df.loc[df['sqft_lot'] == df['sqft_lot'].max()])
print('\n')

print('Home(s) with the least square footage lot:')

#Display data for homes with the square footage equal to the maximum square footage:
print(df.loc[df['sqft_lot'] == df['sqft_lot'].min()])
print('\n')

print('Oldest home(s):')

#Display data for homes with the square footage equal to the maximum square footage:
print(df.loc[df['yr_built'] == df['yr_built'].min()])
print('\n')

print('Newest home(s):')

#Display data for homes with the square footage equal to the maximum square footage:
print(df.loc[df['yr_built'] == df['yr_built'].max()])
print('\n')

print('Home(s) with the most bedrooms:')

#Display data for homes with the most bedrooms:
print(df.loc[df['bedrooms'] == df['bedrooms'].max()])
print('\n')

print('Home(s) with the most bathrooms:')

#Display data for homes with the most bedrooms:
print(df.loc[df['bathrooms'] == df['bathrooms'].max()])
print('\n')
```

```python
#Looking at the descriptions, the variables I am interested in the unique values of are:
#Waterfront - is it truly binary?
#Floors, bedrooms, bathrooms - are there any interesting fractions here?

variables = ['waterfront','floors','bedrooms','bathrooms']

for var in variables :
    print(var,'Unique values:')
    print(df[var].unique())
    print('\n')
```

```python
#Interesting. We know waterfront is a binary variable. Bedrooms are integer values, where
#bathrooms and floors are float values.
#Let's use the groupby command to get a sense of how much more waterfront homes go for.

#Calculate average sales for with/out waterfront
avg_price = df.groupby('waterfront')['price'].mean()

#Reformat for more readability using lambda
avg_price_formatted = avg_price.apply(lambda x: '{:,.2f}'.format(x))

# Print out the results
print("Average Price, With/Without Waterfront")
display(avg_price_formatted)
```

```python
#Looks like on average, a waterfront home sells for about three times as much as one without waterfront.
#Finally, lets use crosstab to view the relationship between bedrooms and floors.

#Creating a bedroom / floor crosstab
tab = pd.crosstab(
    index=df['bedrooms'],
    columns=df['floors'],
)

display(tab)  
```
